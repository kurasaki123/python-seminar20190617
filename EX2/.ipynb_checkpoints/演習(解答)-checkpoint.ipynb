演習１<br>
4都市の折れ線グラフ(気温)を表示<br>
1つのグラフ，横軸は「Month」を指定，タイトルは「Temperature」<br>
できればpd.to_datetime(ヘッダ名)を使って日付データに変換

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data.csv")
header = df.columns.values
df[header[0]] = pd.to_datetime(df[header[0]])

df.plot(x=header[0], y=header[1:], title="Temperature")
plt.show()
```

演習2<br>
4都市の折れ線グラフ(気温)を表示<br>
縦2横2の4グラフ，横軸は「Month」を指定<br>
画像サイズは12, 12

```python
df.plot(x=header[0], y=header[1:], subplots=True, layout=(2, 2), figsize=(12, 12), title="Temperature")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
n = 1;
for ax in axes:
    for a in ax:
        df.plot(ax=a, x=header[0], y=header[n], figsize=(12, 12))
        n += 1;
fig.show()
```

演習3<br>
演習2で作成したグラフのy軸を揃える

```python
df.plot(x=header[0], y=header[1:], subplots=True, layout=(2, 2), figsize=(12, 12), sharey=True)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,sharey=True)
n = 1;
for ax in axes:
    for a in ax:
        df.plot(ax=a, x=header[0], y=header[n], figsize=(12, 12))
        n += 1;
fig.show()
```

演習4<br>
演習3と同じグラフのy軸のラベルを「temperature」にする<br>
各グラフのタイトルを都市名にする

```python
fig, axes = plt.subplots(nrows=2, ncols=2,sharey=True)
n = 1;
for ax in axes:
    for a in ax:
        df.plot(ax=a, x=header[0], y=header[n], figsize=(12, 12), title=header[n])
        n += 1;
        a.set_ylabel("temperature")
fig.show()
```
