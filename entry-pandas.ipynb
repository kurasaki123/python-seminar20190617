```python
df = pd.read_csv("data.csv", names = ["date", "Naha", "Tokyo", "Sapporo"], skiprows = 1, parse_dates = ["date"])
```

```python
#DataFrameを作成
df2 = pd.DataFrame({
    "date":["2019-02-01"],
    "Naha":1,
    "Tokyo":1,
    "Sapporo":3
})
df2
```

```python
#appendメソッドでDataFrame通しの結合
df = df.append(df2)
df
```

```python
#index（行番号）の振り直し
df.reset_index(drop = True)
```

```python
#Naha列の値をつかって並び替え
df.sort_values("Naha", inplace = True)
df
```

```python
#describeメソッドを使ってDataFrameの基本統計情報を表示
df.describe()
```

```python
#Feeling列を追加する。
#Naha列の値が17度以上のFeeling列に「hot」を挿入
#Naha列の値が17度より小さいのFeeling列に「cool」を挿入
df.loc[df["Naha"] >= 17, "Feeling"] = "hot"
df.loc[df["Naha"] < 17, "Feeling"] = "cool"
df
```

```python
#Feelingでグループ化し、hotの値がある行と、coolの値がある行ごとに合計と平均を求める
df.groupby("Feeling").sum()
df.groupby("Feeling").mean()
```
