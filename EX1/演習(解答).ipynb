### 設定
クラスAとクラスBの生徒ごとのテストの点数のデータ（CSV）がある。

### 演習1
2つのクラスのデータを読み込んでそれぞを読み込んで2つのDataFrameをつくる。

```python
import pandas as pd
df_a = pd.read_csv('classA.csv')
df_b = pd.read_csv('classB.csv')
df_a
```

```python
df_b
```

### 演習2
`class`列を追加する。<br/>
classAファイルから作成したDataFrameのclass列には`A`を挿入。<br/>
classBファイルから作成したDataFrameのclass列には`B`を挿入。

```python
df_a["class"] = "A"
df_b["class"] = "B"
df_a
```

```python
df_b
```

### 演習3
2つのDataFrameを合成して1つのDataFrameする。<br>
連番が乱れたら振り直す。

```python
# drop=Trueで古いindexを削除
df = df_a.append(df_b).reset_index(drop=True)
df
```

### 演習4
scoreの高い順にソートする。<br/>

```python
#　ascendingは降順・昇順指定、　inplaceは既存のdfを書き換えるか、新しいdfを返すか
df.sort_values("score", ascending = False, inplace = True)
df
```

### 演習5
`pass/fail`列を追加する。
scoreが７０点以上の生徒の`pass/fail`列には`pass`、７０点より低いの場合は`faill`を挿入。

```python
# 「~」は否定の意味
df.loc[df["score"] >=70, "pass/fail"] = "pass"
df.loc[~(df["score"]>=70), "pass/fail"] = "fail"
df
```

### 演習6
クラスごとの平均、標準偏差、合格者数を求める。<br/>
その際、欠損値があるデータは削除する。

```python
# 欠損値を含むデータの削除
df = df.dropna(how='any') 
```

```python
#平均
df.groupby('class').mean()
```

```python
#標準偏差
df.groupby('class').std()
```

```python
#Aクラス合格者数
len(df[(df["pass/fail"] == "pass") & (df["class"] == "A" )])
```

```python
#Bクラス合格者数
len(df[(df["pass/fail"] == "pass") & (df["class"] == "B" )])
```
