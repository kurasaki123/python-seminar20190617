```python
# pandasとmatplotlibのインポート
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data.csvを読み込んでdataframeの作成
df = pd.read_csv("data.csv", names = ["date", "Naha", "Tokyo", "Sapporo"], skiprows = 1)

# 読み込んだ後でdateを文字列から日付型に変更
df["date"] = pd.to_datetime(df["date"])
df
```

```python
# x軸をdateにしてグラフの作成(y軸には残りの列が使われる)
df.plot(x="date")

# グラフの描画
plt.show()

# グラフの保存
plt.savefig("figure.jpg")
```

```python
# x軸をdate，y軸をTokyo，タイトルをtemperature，グラフサイズを横8，縦6(インチ)
df.plot(x="date", y="Tokyo", title="temperature", figsize=(8, 6))

# グラフの描画
plt.show()
```

```python
# グラフを横に3つ表示
df.plot(x="date", subplots=True, layout=(1, 3), figsize=(18, 6))
```

```python
# 上のグラフのy軸を揃える
df.plot(x="date", subplots=True, sharey=True, layout=(1, 3), figsize=(18, 6))
```

```python
# fig：描画する図，axes：figに描画されるグラフの配列，横3列，縦1行
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(18, 6), sharey=True)

# 0番目(左)のグラフの描画(x軸にはdfのindexが使われる)
df.plot(ax=axes[0], y='Naha')

# 1番目(真ん中)のグラフの描画
df.plot(ax=axes[1], x='date', y='Tokyo')

# 2番目(右)のグラフの描画
df.plot(ax=axes[2], y='Sapporo')

# 0番目のグラフのy軸に名前を付ける
axes[0].set_ylabel("temperature")
```

```python
# fig：描画する図，axes：figに描画されるグラフの配列，横3列，縦1行
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(18, 6), sharey=True)

# 色を赤に
df.plot(ax=axes[0], y='Naha', color='red')

# 点線に
df.plot(ax=axes[1], x='date', y='Tokyo', ls='--')

# マーカーを付ける
df.plot(ax=axes[2], y='Sapporo', marker='.')

axes[0].set_ylabel("temperature")
```

```python
# fig：描画する図，axes：figに描画されるグラフの配列，横3列，縦1行
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(18, 6), sharey=True)

# 棒グラフ
df.plot(ax=axes[0], y='Naha', kind='bar')

# エリアグラフ
df.plot(ax=axes[1], x='date', y='Tokyo', kind='area')

# ヒストグラム
df.plot(ax=axes[2], y='Sapporo', kind='hist')

axes[0].set_ylabel("temperature")
```
